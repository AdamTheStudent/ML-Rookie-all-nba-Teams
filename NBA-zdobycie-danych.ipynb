### Przygotowanie danych zwyciezcow dla ALL-NBA-teams

```python
import pandas as pd
```

```python

df_raw_all_nba = []
lata =[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023,2024]
# https://www.basketball-reference.com/awards/awards_2010.html#leading_all_nba
# https://www.basketball-reference.com/awards/awards_2010.html#all_leading_all_nba
url_general_all_nba = 'https://www.basketball-reference.com/awards/awards_{}.html#all_leading_all_nba'
for rok in lata:
    url = url_general_all_nba.format(rok)
    df = pd.read_html(url, header=1)[2]
    df['season'] = rok    
    # df['IsRookie'] = 
    df_raw_all_nba.append(df)

df_raw_all_nba
```

```python
df_combined = pd.concat(df_raw_all_nba, ignore_index=True)
df_combined = df_combined.dropna(subset=['Age'])#usuniecie wierszy z NaN
# df_combined = df_combined.drop(columns = ["Age", "Pos", "# Tm", "1st Tm", "Pts Won", "Pts Max","2nd Tm", "3rd Tm","Tm", "MP"], axis = 1)
df_combined = df_combined.fillna(0)
df_combined[3:] # w piatym wierszu byly wartosci NaN 
df_combined.to_csv('./csv/all-nba-team.csv')
```

### **Dane wszystkich zawodnikow**

```python
df_raw_all_players = []
lata =[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023,2024]
url_general_all_players =  'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
for rok in lata:
    url = url_general_all_players.format(rok)
    df = pd.read_html(url)[0]
    df['season'] = rok    
    # df['IsRookie'] = 
    df_raw_all_players.append(df)

df_raw_all_players
```

```python
df_combined_players = pd.concat(df_raw_all_players, ignore_index=True)
# df_combined.to_csv('./csv/players.csv') # mozliwosc zapisu danych 
df_combined_players[22:] #rzad 25 i nastepne nie zawieraja danych zawodnikow
```

```python
df_combined_players = df_combined_players.drop(df_combined_players[df_combined_players.Age == 'Age'].index)
df_combined_players[22:]
```

```python
df_combined_players.to_csv('./csv/players.csv')
```
