## Stworzenie modelu predykcji dla all-NBA-team oraz all-rookie-team

### Czyszczenie danych dla all-nba-team

```python
import pandas as pd
```

```python
all_nba = pd.read_csv("./csv/all-nba-team.csv")
players = pd.read_csv("./csv/players.csv")
rookies = pd.read_csv('./csv/rookies.csv')
all_rookies = pd.read_csv('./csv/rookies_voting.csv')
```

```python
all_nba = all_nba[["Player","season","Pts Won","Pts Max","Share"]]
all_nba
```

```python
players = players.drop(["Unnamed: 0","Rk"], axis=1)
# players
```

```python
players["Player"]=players["Player"].str.replace("*","",regex=False)
```

```python
def usun_duplikaty(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
        
players = players.groupby(["Player","season"]).apply(usun_duplikaty)
players.index = players.index.droplevel()
players.index = players.index.droplevel()
```

```python
players.head(10)
```

```python
combined_all_nba = players.merge(all_nba, how = "outer", on = ["Player","season"])
combined_all_nba.head(10)
```

```python
combined_all_nba[["Pts Won", "Pts Max", "Share"]] = combined_all_nba[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined_all_nba = combined_all_nba.apply(pd.to_numeric, errors = "ignore")
```

```python
combined_all_nba.dtypes
```

```python
combined_all_nba.to_csv("./csv/all_nba_cleaned.csv")
```

### Czyszczenie danych dla all-rookie

```python
all_rookie = all_rookies[["Player","season","Pts Won","Pts Max","Share"]]
all_rookie
```

```python
rookies
```

```python
rookies = rookies.drop(["Unnamed: 0","Rk"], axis=1)
```

```python
rookies["Player"]=rookies["Player"].str.replace("*","",regex=False)
```

```python
combined_rookies = rookies.merge(all_rookie, how = "outer", on = ["Player","season"])
combined_rookies.head(20)
```

```python
combined_rookies[["Pts Won", "Pts Max", "Share"]] = combined_rookies[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined_rookies = combined_rookies.apply(pd.to_numeric, errors = "ignore")
combined_rookies
```

```python
combined_rookies.dtypes
```

```python
combined_rookies.to_csv("./csv/all_rookies_cleaned.csv")
```
